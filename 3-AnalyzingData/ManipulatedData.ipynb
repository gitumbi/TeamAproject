```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
import seaborn as sns
%matplotlib inline
import  statsmodels.api as sm
import statsmodels.stats.api as sms
from scipy.stats import boxcox
```

```python
Man_data = pd.read_csv ("Merged_data.csv")
print (Man_data)
```

```python
##US$1000M
```

```python
sns.distplot(Man_data['US$1000M'])
```

```python
Man_data['US$1000MSQ'] = Man_data['US$1000M']**2
```

```python
sns.distplot(Man_data['US$1000MSQ'])
```

```python
Man_data['US$1000MCUBE'] = Man_data['US$1000M']**3
```

```python
sns.distplot(Man_data['US$1000MCUBE'])
```

```python
x = Man_data['TempChangeM']
y = Man_data['US$1000MCUBE']
model = sm.OLS(y,x).fit()
```

```python
pred_val = model.fittedvalues.copy()
true_val = Man_data['US$1000MCUBE'].values.copy()
residual = true_val - pred_val
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
_ = ax.scatter(residual, pred_val)
```
