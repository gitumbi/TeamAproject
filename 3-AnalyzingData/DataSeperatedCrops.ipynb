```python
##Splitting the dataset by crop
```

```python
# Import packages and set visualization style
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
Split_data = pd.read_csv('Recode_data.csv')
Split_data.head()
```

```python
##Checking for null
```

```python
Split_data.info()
```

```python
##Splitting the dataframe by crop type
```

```python
df1, df2, df3, = [x for _, x in Split_data.groupby(Split_data['Crops'])]
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
##Renaming columns they have in common we don't want to Merge by
```

```python
df1.rename(columns = {'Crops':'CropsM', 'Tonnes':'TonnesM',
                              'US$1000':'US$1000M', 'TempChange':'TempChangeM', 'CropsR':'CropsRM'}, inplace = True)
```

```python
df1.head()
```

```python
df2.rename(columns = {'Crops':'CropsSB', 'Tonnes':'TonnesSB',
                              'US$1000':'US$1000SB', 'TempChange':'TempChangeSB', 'CropsR':'CropsRSB'}, inplace = True)
```

```python
df2.head()
```

```python
df3.rename(columns = {'Crops':'CropsW', 'Tonnes':'TonnesW',
                              'US$1000':'US$1000W', 'TempChange':'TempChangeW', 'CropsR':'CropsRW'}, inplace = True)
```

```python
df3.head()
```

```python
##Merging Dataframes back together on Year and YearR
```

```python
from functools import reduce
```

```python
dfs = [df1, df2, df3]
```

```python
Merged_data = reduce(lambda left,right: pd.merge(left,right,on='YearR'), dfs)
```

```python
Merged_data.head()
```

```python
##Removing Columns
```

```python
Merged_data.drop(['Unnamed: 0_x', 'Year_x', 'Unnamed: 0_y', 'Year_y', 'Unnamed: 0'], axis = 1, inplace=True)
```

```python
Merged_data.head()
```

```python
Merged_data.to_csv('Merged_data.csv')
```
